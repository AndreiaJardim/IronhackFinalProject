```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
from selenium.common.exceptions import NoSuchElementException
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import time

# Set the Edge WebDriver executable path
edge_driver_path = 'C:\\Users\\jardi\\Desktop\\Ironhack\\FINAL PROJECT\\EgdeDriver\\msedgedriver.exe'

# Set the language preference to English
options = webdriver.EdgeOptions()
options.add_argument('--lang=en-US')

# Initialize the Edge driver
driver = webdriver.Edge(executable_path=edge_driver_path, options=options)

# Create a DataFrame to store the wine data
winedata_list = []

# Iterate through Wine Links in red_wine DataFrame
for wine_link in red_wine['Wine Link']:
    # Build the URL
    full_url = f"https://www.vivino.com{wine_link}"

    # Navigate to the wine page
    driver.get(full_url)

    try:
        # Check if cookies exist
        cookies_button = driver.find_element(By.XPATH, '//*[@id="cookie-notice-container"]/div/button')
        cookies_button.click()
        time.sleep(2)
    except NoSuchElementException:
        pass  # The cookie notice was not found, continue without clicking it

    # Extract wine information (ID, name, winery)
    page_source = driver.page_source
    soup = BeautifulSoup(page_source, 'html.parser')
    wine_id = soup.find('a', class_='wine')['href'].split('/')[-1]
    wine_name = soup.find('a', class_='wine').text.strip()
    winery = soup.find('a', class_='winery').text.strip()

    # Store the current window handle (main page)
    main_page = driver.current_window_handle

    # Slow scrolling
    total_height = int(driver.execute_script("return document.body.scrollHeight"))
    for i in np.linspace(1, total_height * 2, 7):
        driver.execute_script("window.scrollTo(0, {});".format(i))
        time.sleep(2)

    # Check if the element "Show all" exists
    if driver.find_elements(By.XPATH, '//*[@id="wine-page-lower-section"]/div[3]/div/div[2]/a'):
        element = driver.find_element(By.XPATH, '//*[@id="wine-page-lower-section"]/div[3]/div/div[2]/a')
        time.sleep(1)
        actions = ActionChains(driver)
        actions.move_to_element(element).perform()

        try:
            show_all_button = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.XPATH, '//*[@id="wine-page-lower-section"]/div[3]/div/div[2]/a'))
            )
            show_all_button.click()
            time.sleep(5)  # Wait for the pop-up window to load

            # Switch to the new pop-up window
            for handle in driver.window_handles:
                if handle != main_page:
                    popup = handle
                    driver.switch_to.window(popup)

            # Find the vintage elements within the pop-up window
            vintage_elements = driver.find_elements(By.CLASS_NAME, 'vintageListRow__vintageListRow--1wm0Y')

            # Loop through the vintage elements to extract information
            for vintage_element in vintage_elements:
                year = vintage_element.find_element(By.CLASS_NAME, 'vintageListRow__year--34Tuc').text.strip()
                average_rating = vintage_element.find_element(By.CLASS_NAME, 'vivinoRating_averageValue__uDdPM').text.strip()
                ratings = vintage_element.find_element(By.CLASS_NAME, 'vivinoRating_caption__xL84P').text.strip()
                price = vintage_element.find_element(By.CLASS_NAME, 'vintageListRow__wrapper--3AGrd').text.strip()

                # Add the extracted information to the data list
                winedata_list.append({
                    'Wine id': wine_id,
                    'Wine Name': wine_name,
                    'Winery': winery,
                    'Year': year,
                    'Average Rating': average_rating,
                    'Ratings': ratings,
                    'Price': price
                })

            # Close the pop-up window
            driver.close()

            # Switch back to the main window
            driver.switch_to.window(main_page)

        except NoSuchElementException:
            # Handle the case where the pop-up window or its elements are not found
            pass

    # Check if the vintage section element is found
    elif driver.find_elements(By.XPATH, '//*[@id="vintageListSection"]'):
        element = driver.find_element(By.XPATH, '//*[@id="vintageListSection"]')
        time.sleep(1)
        actions = ActionChains(driver)
        actions.move_to_element(element).perform()

        # Find the vintage elements 
        vintage_elements = driver.find_elements(By.CLASS_NAME, 'vintageListRow__vintageListRow--1wm0Y')

        # Loop through the vintage elements to extract information
        for vintage_element in vintage_elements:
            year = vintage_element.find_element(By.CLASS_NAME, 'vintageListRow__year--34Tuc').text.strip()
    
            # Find all elements that contain the average rating
            average_rating_elements = vintage_element.find_elements(By.CSS_SELECTOR, '.vivinoRating_averageValue__uDdPM')
            average_ratings = [element.text.strip() for element in average_rating_elements]
            average_rating = ", ".join(average_ratings) if average_ratings else "No average rating available"

            # Find all elements that contain the ratings
            rating_elements = vintage_element.find_elements(By.CSS_SELECTOR, '#vintageListSection div.vivinoRating_caption__xL84P')
            ratings = [element.text.strip() for element in rating_elements]
            ratings = ", ".join(ratings) if ratings else "No ratings available"

            # Find all elements that contain the price
            price_elements = vintage_element.find_elements(By.CSS_SELECTOR, '#vintageListSection div.col.mobile-column-6.tablet-column-12.desktop-column-3')
            prices = [element.text.strip() for element in price_elements]
            price = ", ".join(prices) if prices else "No average price available"

            # Add the extracted information to the data list
            winedata_list.append({
                'Wine id': wine_id,
                'Wine Name': wine_name,
                'Winery': winery,
                'Year': year,
                'Average Rating': average_rating,
                'Ratings': ratings,
                'Price': price
            })

    else:
        # Handle the case where the vintage information is displayed differently
        vintage_element = soup.find('span', class_='vintage')
        year = vintage_element.text.strip()
        average_rating = driver.find_element(By.CLASS_NAME, 'vivinoRating_averageValue__uDdPM').text.strip()
        ratings = driver.find_element(By.CLASS_NAME, 'vivinoRating_caption__xL84P').text.strip()
        try:
            price_element = driver.find_element(By.CSS_SELECTOR, 'span.purchaseAvailabilityPPC__amount--2_4GT')
            price = price_element.text.strip()
        except NoSuchElementException:
            price = "No average price available"


        # Add the extracted information to the data list
        winedata_list.append({
            'Wine id': wine_id,
            'Wine Name': wine_name,
            'Winery': winery,
            'Year': year,
            'Average Rating': average_rating,
            'Ratings': ratings,
            'Price': price
        })

# Create a DataFrame from the extracted data
red_winesdata = pd.DataFrame(winedata_list)

# Print the DataFrame
print(red_winesdata)

# Close the browser window
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
from selenium.common.exceptions import NoSuchElementException
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import time

# Set the Edge WebDriver executable path
edge_driver_path = 'C:\\Users\\jardi\\Desktop\\Ironhack\\FINAL PROJECT\\EgdeDriver\\msedgedriver.exe'

# Set the language preference to English
options = webdriver.EdgeOptions()
options.add_argument('--lang=en-US')

# Initialize the Edge driver
driver = webdriver.Edge(executable_path=edge_driver_path, options=options)

# Create a DataFrame to store the wine data
winedata_list = []

# Iterate through Wine Links in red_wine DataFrame
for wine_link in fortified_wine['Wine Link']:
    # Build the URL
    full_url = f"https://www.vivino.com{wine_link}"

    # Navigate to the wine page
    driver.get(full_url)

    try:
        # Check if cookies exist
        cookies_button = driver.find_element(By.XPATH, '//*[@id="cookie-notice-container"]/div/button')
        cookies_button.click()
        time.sleep(2)
    except NoSuchElementException:
        pass  # The cookie notice was not found, continue without clicking it

    # Extract wine information (ID, name, winery)
    page_source = driver.page_source
    soup = BeautifulSoup(page_source, 'html.parser')
    wine_id = soup.find('a', class_='wine')['href'].split('/')[-1]
    wine_name = soup.find('a', class_='wine').text.strip()
    winery = soup.find('a', class_='winery').text.strip()

    # Store the current window handle (main page)
    main_page = driver.current_window_handle

    # Slow scrolling
    total_height = int(driver.execute_script("return document.body.scrollHeight"))
    for i in np.linspace(1, total_height * 2, 7):
        driver.execute_script("window.scrollTo(0, {});".format(i))
        time.sleep(2)

    # Check if the element "Show all" exists
    if driver.find_elements(By.XPATH, '//*[@id="wine-page-lower-section"]/div[3]/div/div[2]/a'):
        element = driver.find_element(By.XPATH, '//*[@id="wine-page-lower-section"]/div[3]/div/div[2]/a')
        time.sleep(1)
        actions = ActionChains(driver)
        actions.move_to_element(element).perform()

        try:
            show_all_button = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.XPATH, '//*[@id="wine-page-lower-section"]/div[3]/div/div[2]/a'))
            )
            show_all_button.click()
            time.sleep(5)  # Wait for the pop-up window to load

            # Switch to the new pop-up window
            for handle in driver.window_handles:
                if handle != main_page:
                    popup = handle
                    driver.switch_to.window(popup)

            # Find the vintage elements within the pop-up window
            vintage_elements = driver.find_elements(By.CLASS_NAME, 'vintageListRow__vintageListRow--1wm0Y')

            # Loop through the vintage elements to extract information
            for vintage_element in vintage_elements:
                year = vintage_element.find_element(By.CLASS_NAME, 'vintageListRow__year--34Tuc').text.strip()
                average_rating = vintage_element.find_element(By.CLASS_NAME, 'vivinoRating_averageValue__uDdPM').text.strip()
                ratings = vintage_element.find_element(By.CLASS_NAME, 'vivinoRating_caption__xL84P').text.strip()
                price = vintage_element.find_element(By.CLASS_NAME, 'vintageListRow__wrapper--3AGrd').text.strip()

                # Add the extracted information to the data list
                winedata_list.append({
                    'Wine id': wine_id,
                    'Wine Name': wine_name,
                    'Winery': winery,
                    'Year': year,
                    'Average Rating': average_rating,
                    'Ratings': ratings,
                    'Price': price
                })

            # Close the pop-up window
            driver.close()

            # Switch back to the main window
            driver.switch_to.window(main_page)

        except NoSuchElementException:
            # Handle the case where the pop-up window or its elements are not found
            pass

    # Check if the vintage section element is found
    elif driver.find_elements(By.XPATH, '//*[@id="vintageListSection"]'):
        element = driver.find_element(By.XPATH, '//*[@id="vintageListSection"]')
        time.sleep(1)
        actions = ActionChains(driver)
        actions.move_to_element(element).perform()

        # Find the vintage elements 
        vintage_elements = driver.find_elements(By.CLASS_NAME, 'vintageListRow__vintageListRow--1wm0Y')

        # Loop through the vintage elements to extract information
        for vintage_element in vintage_elements:
            year = vintage_element.find_element(By.CLASS_NAME, 'vintageListRow__year--34Tuc').text.strip()
    
            # Find all elements that contain the average rating
            average_rating_elements = vintage_element.find_elements(By.CSS_SELECTOR, '.vivinoRating_averageValue__uDdPM')
            average_ratings = [element.text.strip() for element in average_rating_elements]
            average_rating = ", ".join(average_ratings) if average_ratings else "No average rating available"

            # Find all elements that contain the ratings
            rating_elements = vintage_element.find_elements(By.CSS_SELECTOR, '#vintageListSection div.vivinoRating_caption__xL84P')
            ratings = [element.text.strip() for element in rating_elements]
            ratings = ", ".join(ratings) if ratings else "No ratings available"

            # Find all elements that contain the price
            price_elements = vintage_element.find_elements(By.CSS_SELECTOR, '#vintageListSection div.col.mobile-column-6.tablet-column-12.desktop-column-3')
            prices = [element.text.strip() for element in price_elements]
            price = ", ".join(prices) if prices else "No average price available"

            # Add the extracted information to the data list
            winedata_list.append({
                'Wine id': wine_id,
                'Wine Name': wine_name,
                'Winery': winery,
                'Year': year,
                'Average Rating': average_rating,
                'Ratings': ratings,
                'Price': price
            })

    else:
        # Handle the case where the vintage information is displayed differently
        vintage_element = soup.find('span', class_='vintage')
        year = vintage_element.text.strip()
        average_rating = driver.find_element(By.CLASS_NAME, 'vivinoRating_averageValue__uDdPM').text.strip()
        ratings = driver.find_element(By.CLASS_NAME, 'vivinoRating_caption__xL84P').text.strip()
        try:
            price_element = driver.find_element(By.CSS_SELECTOR, 'span.purchaseAvailabilityPPC__amount--2_4GT')
            price = price_element.text.strip()
        except NoSuchElementException:
            price = "No average price available"


        # Add the extracted information to the data list
        winedata_list.append({
            'Wine id': wine_id,
            'Wine Name': wine_name,
            'Winery': winery,
            'Year': year,
            'Average Rating': average_rating,
            'Ratings': ratings,
            'Price': price
        })

# Create a DataFrame from the extracted data
fortified_winesdata = pd.DataFrame(winedata_list)

# Print the DataFrame
print(fortified_winesdata)

# Close the browser window
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import time

from selenium.webdriver.edge.service import Service as EdgeService
# Set the Edge WebDriver executable path
edge_driver_path = 'C:\\Users\\jardi\\Desktop\\Ironhack\\FINAL PROJECT\\EgdeDriver\\msedgedriver.exe'
# Set the language preference to English
options = webdriver.EdgeOptions()
options.add_argument('--lang=en-US')

# Initialize the Edge service
edge_service = EdgeService(executable_path=edge_driver_path)

# Initialize the Edge driver with the service object
driver = webdriver.Edge(service=edge_service, options=options)

# Create a DataFrame to store the wine data
winedata_list = []

try:
    # Check if cookies exist
    cookies_button = driver.find_element(By.XPATH, '//*[@id="cookie-notice-container"]/div/button')
    cookies_button.click()
    time.sleep(2)
except NoSuchElementException:
    pass  # The cookie notice was not found, continue without clicking it

# Iterate through Wine Links in wine_df1 DataFrame
for wine_link in wine_list_ratings['Short Wine Link']:
    try:
        # Build the URL
        full_url = f"https://www.vivino.com{wine_link}"

        # Navigate to the wine page
        driver.get(full_url)

        # Extract wine information (ID, name, winery)
        page_source = driver.page_source
        soup = BeautifulSoup(page_source, 'html.parser')
        wine_id = soup.find('a', class_='wine')['href'].split('/')[-1]
        wine_name = soup.find('a', class_='wine').text.strip()
        winery = soup.find('a', class_='winery').text.strip()
        image_element = driver.find_element(By.CLASS_NAME, 'image')
        image_url = image_element.get_attribute('src')
        average_price_element = soup.find('span', class_='purchaseAvailabilityPPC__amount--2_4GT')
        average_price = average_price_element.text.strip() if average_price_element else None

        # Slow scrolling
        total_height = int(driver.execute_script("return document.body.scrollHeight"))
        for i in np.linspace(1, total_height * 2, 7):
            driver.execute_script("window.scrollTo(0, {});".format(i))
            time.sleep(2)

        try:
            # Use WebDriverWait to wait for the table element with a dynamic selector
            table = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.CSS_SELECTOR, '#wine-page-lower-section > div.grid.vintagePage__bottom--1fcqg > div > div:nth-child(4) > table'))
            )

            # Now you can work with the 'table' element
            rows = table.find_elements(By.XPATH, './tbody/tr')

            # Initialize variables to accumulate values
            grapes_text = ""
            region_text = ""
            wine_style_text = ""
            wine_style_text_element = ""

            # Process the table rows
            for row in rows:
                label = row.find_element(By.XPATH, './th/div/span[2]')
                if label:
                    label_text = label.text.strip()
                    if label_text == "Allergens" or label_text == "Alcohol content":
                        specific_element = row.find_element(By.XPATH, './td/span')
                        text = specific_element.text.strip()
                        #print("Label:", label_text)
                        #print("Text:", text)
                    else:
                        links = row.find_elements(By.XPATH, './td/a')
                        href_list = [link.get_attribute("href") for link in links]
                        href_string = ', '.join(href_list)

                        # Extract the text element within the same cell
                        text_elements = row.find_elements(By.XPATH, './td/a')
                        text_element_texts = [text_element.text.strip() for text_element in text_elements]
                        text_element_text = ', '.join(text_element_texts)

                        # Print both href and text
                        #print("Label:", label_text)
                        #print("Hrefs:", href_string)
                        #print("Text:", text_element_text)

                        # Accumulate values for Grapes, Region, and Wine Style
                        if label_text == "Grapes":
                            grapes_text = href_string
                            grapes_text_element = text_element_text
                        elif label_text == "Region":
                            region_text = href_string
                            region_text_element = text_element_text
                        elif label_text == "Wine style":
                            wine_style_text = href_string
                            wine_style_text_element = text_element_text

            # Extracted information to be added to the data list
            wine_data = {
                'Wine id': wine_id,
                'Wine Name': wine_name,
                'Winery': winery,
                'ImageURL': image_url,
                'Grapes_url': grapes_text,
                'Region_url': region_text,
                'Wine Style_ur': wine_style_text,
                'Grapes': grapes_text_element,
                'Region': region_text_element,
                'Wine Style': wine_style_text_element,
                'Alcohol Content': text if label_text == "Alcohol content" else "N/A",
                'Allergens': text if label_text == "Allergens" else "N/A",
                'Average Price': average_price
            }

            # Append the extracted data to the list
            winedata_list.append(wine_data)

            pass

        except Exception as e:
            print(f"An error occurred: {str(e)}")

    except Exception as e:
        print(f"An error occurred: {str(e)}")

# Create a DataFrame from the extracted data
wine_detail = pd.DataFrame(winedata_list)

# Export to a new CSV file
wine_detail.to_csv('wine_characteristics_data.csv', index=False) 

# Make sure to close the WebDriver when done
driver.quit()
```
